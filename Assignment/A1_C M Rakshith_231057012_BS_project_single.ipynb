```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# URL choosen for scraping
URL = "https://www.goodreads.com/genres/novels"

# Send a GET request to the website
response = requests.get(URL)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
# Creating a list
novel_details = []

# Find all the div elements with the specified class
novel_blocks = soup.find_all('div', class_='coverWrapper')

# Extract and store the text in the novel_details list
for block in novel_blocks:
    details = {}
    title = block.find('img')['alt']
    details['Title'] = title

    novel_details.append(details)
print(novel_details)
```

```python
# Creation of CSV file
filename = 'NovelTitle.csv'
with open(filename, 'w', newline='') as f:
    fieldnames = ['Title']
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(novel_details)
```
